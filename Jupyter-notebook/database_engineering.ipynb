```python
# add dependencies
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

from sqlalchemy import Column, Integer, String, Float, Date
```

```python
# create the avenger2012_data class
class Avenger2012_data(Base):
    __tablename__ = 'avenger2012_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the avenger2015_data class
class Avenger2015_data(Base):
    __tablename__ = 'avenger2015_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the avenger2018_data class
class Avenger2018_data(Base):
    __tablename__ = 'avenger2018_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the country class
class Country(Base):
    __tablename__ = 'country'
    index = Column(Integer, primary_key = True)
    country = Column(String(255))
    latitude = Column(String(255))
    longitude = Column(String(255))
    name = Column(String(255))
    
# create the studio_market_share class
class Studio_market_share(Base):
    __tablename__ = 'studio_market_share'
    index = Column(Integer, primary_key = True)
    year_id = Column(Float)
    Studio_name = Column(String(255))
    Studio_Gross = Column(Float)
    Total_Yearly_Gross = Column(Float)
    Percentage = Column(Float)
    pop_size = Column(Float)
    
# create the studio_market_share class
class Box_office(Base):
    __tablename__ = 'box_office'
    index = Column(Integer, primary_key = True)
    Studio = Column(String(255))
    Year = Column(Float)
    Total_Gross = Column(Float)
    box_office = Column(Float)

# create the final_animation_data class
class Final_animation_data(Base):
    __tablename__ = 'final_animation_data'
    index = Column(Integer, primary_key = True)
    Studio = Column(String(255))
    Year = Column(Float)
    Genre1 = Column(String(255))
    Total_Gross = Column(Float)
    RottenTomato = Column(Float)
    
# create the oscar class
class Oscar(Base):
    __tablename__ = 'oscar'
    index = Column(Integer, primary_key = True)
    Studio = Column(String(255))
    Year = Column(Float)
    Oscar_winner = Column(Float)

# create the Rotten_Tomato class
class Rotten_Tomato(Base):
    __tablename__ = 'rotten_tomato'
    index = Column(Integer, primary_key = True)
    Studio = Column(String(255))
    RottenTomato = Column(Float)
# create the oscars overall class    
class Oscar_Overall(Base):
    __tablename__ = 'oscar_all'
    index = Column(Integer, primary_key = True)
    Studio = Column(String(255))
    RottenTomato = Column(Float)    
```

```python
# create the datbase and add the classes
engine = create_engine("sqlite:///database.sqlite", echo = False)
conn = engine.connect()
session = Session(engine)

Base.metadata.create_all(conn)
```
