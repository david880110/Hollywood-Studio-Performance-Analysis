```python
# add dependencies
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

from sqlalchemy import Column, Integer, String, Float, Date
```

```python
# create the avenger2012_data class
class Avenger2012_data(Base):
    __tablename__ = 'avenger2012_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the avenger2015_data class
class Avenger2015_data(Base):
    __tablename__ = 'avenger2015_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the avenger2018_data class
class Avenger2018_data(Base):
    __tablename__ = 'avenger2018_data'
    index = Column(Integer, primary_key = True)
    Country = Column(String(255))
    Dist = Column(String(255))
    Release = Column(Date)
    Opening = Column(Float)
    percent_of = Column(Float)
    Total_Gross = Column(Float)
    As_Of = Column(Date)
    
# create the country class
class Country(Base):
    __tablename__ = 'country'
    index = Column(Integer, primary_key = True)
    country = Column(String(255))
    latitude = Column(String(255))
    longitude = Column(String(255))
    name = Column(String(255))
```

```python
# create the datbase and add the classes
engine = create_engine("sqlite:///avenger.sqlite", echo = False)
conn = engine.connect()
session = Session(engine)

Base.metadata.create_all(conn)
```

```python
# Load csv files
first_file = 'cleandata_SQL/avenger2012_data.csv'
second_file = 'cleandata_SQL/avenger2015_data.csv'
third_file = 'cleandata_SQL/avenger2018_data.csv'
c_file = 'cleandata_SQL/countries.csv'
```

```python
# Read 1st file with pandas
first_df = pd.read_csv(first_file)
first_df.head()
```

```python
# Read 2nd file with pandas
second_df = pd.read_csv(second_file)
second_df.head()
```

```python
# Read 3rd file with pandas
third_df = pd.read_csv(third_file)
third_df.head()
```

```python
# Read c file with pandas
c_df = pd.read_csv(c_file)
c_df.head()
```

```python
# send the station dataframe to the sqlite database
first_df.to_sql('avenger2012_data', engine, if_exists='replace')
second_df.to_sql('avenger2015_data', engine, if_exists='replace')
third_df.to_sql('avenger2018_data', engine, if_exists='replace')
c_df.to_sql('country', engine, if_exists='replace')
```
